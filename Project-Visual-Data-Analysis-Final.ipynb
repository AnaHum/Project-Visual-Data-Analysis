```python
#Step1-2
#Load the csv and excel files into pandas.limit the amount of rows


import pandas as pd
import numpy as np
from IPython.display import Image
import os

fert = pd.read_csv('gapminder_total_fertility.csv', index_col=0, nrows=260)
life = pd.read_excel('gapminder_lifeexpectancy.xlsx.', index_col=0, nrows=260)
pop = pd.read_excel('gapminder_population.xlsx.', index_col=0, nrows=260)
```

```python
#Step 3
#Check the shape of all the tables (rows, columns):


life.shape, fert.shape, pop.shape
```

```python
#Check the Summerization of the tables

life.describe, fert.describe, pop.describe
```

```python
#Step4
#Take a look at the columns

life.columns, fert.columns, pop.columns
```

```python
#Convert fert table into the same format as life and pop so integer:

fert.columns = fert.columns.astype(int)
fert.columns
```

```python
#Step 5
#Check the row index of the fert table and name the index as 'country'

fert.index
fert.index.name = 'country'
fert
```

```python
#Step 6
# Move the row index into a column:

fert = fert.reset_index()
fert
```

```python
#Convert the table into long format. 

fert = fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')
fert
```

```python
#Step7
#Check the row index of the fert table and name the index as 'country'

life.index
life.index.name = 'country'
life
```

```python
#Move the row index into a column.
#Convert the table into long format. 

life = life.reset_index()
life = life.melt(id_vars='country', var_name='year', value_name='life_expectancy')
life
```

```python
#Step 8
#Check the row index of the fert table and name the index as 'country'

pop.index
pop.index.name = 'country'
pop
```

```python
#Move the row index into a column.
#Convert the table into long format. 

pop = pop.reset_index()
pop = pop.melt(id_vars='country', var_name='year', value_name='population')
pop
```

```python
#Step 9
#Merge the DataFrames together. 

df = fert.merge(pop)
df
```

```python
#Merge the data with the third table

df = df.merge(life)
df
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
import csv
```

```python
#Select specific columns from a final merged df and plot them:

df_subset = df.loc[df['country'].isin(['France', 'Germany', 'Sweden'])]
sns.lineplot(x='year', y='life_expectancy', hue='country', data=df_subset)
```

```python
# A scatterplot of a an above selected columns:

sns.scatterplot(x='year', y='life_expectancy', hue='country',
            data=df_subset, alpha=0.6)
```

```python
# limiting years and countries in a cdf and clearing the data

limited1 = df [ df ['year'] > 1959]
limited2 = limited1 [ limited1 ['year'] <2016]
limited2= limited2 [limited2['country'].isin(['France','Portugal', 'Germany', 'Norway','Denmark','Belgium','Italy','Greece','Spain','Sweden','Norway','Austria','Finland','Netherlands'])].dropna()  
limited2


```

```python
#Create one plot for each year between 1960 and 2015 
#Write a for loop that goes through each year and produces a scatterplot.
#Save each scatterplot to a separate file containing the year in the filename, e.g. lifeexp_1999.png.
#After saving the plot call plt.close() to remove the plot from the notebook.
#Save each scatterplot to a separate file containing the year in the filename and remove the plot from the notebook.


sy=1960
ey=2016

for year in range(sy,ey):
    ndf1 = limited2.loc[limited2['year']==year]
    sns.scatterplot(x='life_expectancy', y='fertility_rate', hue='country', size='year', data=ndf1,  alpha=0.8)
    plt.axis((60, 85, 1, 4))
    plt.savefig(f'lifeexp_{year}.png', dpi=200) 
    plt.close() 
    
    
    
    
```

```python
#Step 12
#Add another dimension to the visualization by utilizing the population data.
#Adjust the plot, so that the size of the symbols represent the population from the gapminder dataset.
#Save each scatterplot to a separate file containing the year in the filename and remove the plot from the notebook.

for year in range(sy,ey):
    ndf2 = limited2.loc[limited2['year']==year]
    sns.scatterplot(x='life_expectancy', y='fertility_rate', hue='country', size='population', 
                    data=ndf2,
                    alpha=0.8)

    plt.axis((60, 85, 1, 5))
    plt.savefig(f'pop_{year}.png', dpi=200)
    plt.close()
    
    
    
```

```python
#Alternative way - Using Lmplot function

for year in range(sy, ey):
    ndf3 = limited2.loc[limited2['year']==year]
    sns.lmplot(x='life_expectancy', y='fertility_rate',  data= ndf3 , height = 5, col = 'year' , ci= None, line_kws= {'color':'red'}, scatter_kws = {'color' :'green'}, col_wrap = 3).set(title='The correlation between life expectancy and fertility in '+ str(year))
    plt.show()
    

    
```

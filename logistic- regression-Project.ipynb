```python
#step1
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
#step2

df = pd.read_csv('train.csv')
df.head()
```

```python
df.drop('Cabin',axis =1,inplace=True)
```

```python
df.head()
```

```python
df['Age'].isnull().sum() ,df['Pclass'].isnull().sum(), df['Sex'].isnull().sum(), df['SibSp'].isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace = True)
```

```python
df['Age'].isnull().sum()
```

```python
df.head()
```

```python
df['Sex']=df ['Sex'].replace(['male'], '0')

df['Sex']=df ['Sex'].replace(['female'], '1')
```

```python

df.head()
```

```python
#Select columns for y and X

X = df[['Pclass', 'Age', 'Sex', 'SibSp']] 
y=df['Survived']
```

```python
X

```

```python
y
```

```python
y.shape, X.shape
```

```python
# Split the DataFrame into X and y
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
#Step 4

sns.scatterplot(data=X_train, x='Age', y='Pclass', hue=y_train)

```

```python
#Step 5

df['Age'].isnull().sum()

```

```python
X
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression() #initialize the model
model.fit(X_train, y_train)   # trains the model:
# the LogReg looks at the data, determines the model parameters that position the 
# decision boundary in such a way that separates the training data best
model.score(X_train, y_train) # calculates accuracy

```

```python
model.score(X_test, y_test)
# you want the training + test score to be very similar

```
